```python
import pandas as pd
```

```python
df = pd.read_csv('data/loan.csv')
df.head()
df.dropna(axis=1, how='all', inplace=True)
df.dropna(axis=0, how='all', inplace=True)
df.head()
```

```python
dic = pd.read_excel('data/LCDataDictionary.xlsx')
```

```python
data = df.drop(["collection_recovery_fee","emp_title","funded_amnt_inv","installment","last_credit_pull_d",
                "last_pymnt_amnt","last_pymnt_d","loan_amnt","member_id","next_pymnt_d","out_prncp","out_prncp_inv","recoveries",
                "total_pymnt","total_pymnt_inv","total_rec_int","total_rec_late_fee","total_rec_prncp","url","zip_code","desc",
                "title","addr_state","policy_code","issue_d"], axis=1)
```

```python
data = data.drop(data[(data['loan_status'] == 'Current') | (data['loan_status'] == 'Issued')].index)
data['loan_status'].value_counts()
```

```python
good_loan = ((data['loan_status'] == 'Fully Paid') |  (data['loan_status'] == 'Does not meet the credit policy. Status:Fully Paid')).astype(int)
data['good_loan'] = good_loan
data.head()
```

```python
data = data.drop('loan_status', axis=1)
data.to_csv('data/variables.csv')
data.head()
```
